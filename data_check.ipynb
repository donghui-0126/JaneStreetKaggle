```python
symbol_set = set()
symbol_num = {}

for i in range(0,10):
   temp = pd.read_parquet(f"/home/luke/pistar/jane-street-real-time-market-data-forecasting/train.parquet/partition_id={i}/part-0.parquet")
   symbol_set.update(temp.symbol_id.unique().tolist())
   
   # symbol_id별 count를 계산
   counts = temp.symbol_id.value_counts()
   
   # 각 symbol_id의 count를 누적
   for symbol_id, count in counts.items():
       if symbol_id in symbol_num:
           symbol_num[symbol_id] += count
       else:
           symbol_num[symbol_id] = count
```
